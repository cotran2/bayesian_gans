```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
generated = np.load('generated_sample.npy')
random = np.load('random_sample.npy')
```

```python
generated
```

```python
plt.hexbin(generated[:,0], generated[:,1], cmap='rainbow')
plt.gca().set_aspect('equal', adjustable='box')
plt.xlim([-3, 3])
plt.ylim([-3, 3])
plt.show()
```

```python
plt.hexbin(random[:,0], random[:,1], cmap='rainbow')
plt.gca().set_aspect('equal', adjustable='box')
plt.xlim([-3, 3])
plt.ylim([-3, 3])
plt.show()
```
